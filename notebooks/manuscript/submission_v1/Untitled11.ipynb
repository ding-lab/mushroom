```python
SEQUENTIAL_CMAPS = ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                      'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                      'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']
def get_hierarchical_cmap(label_to_hierarchy):
    aggs = np.stack(label_to_hierarchy.values())
    n_clusts = aggs.max(0)
    n_maps = n_clusts[0] + 1
        
    print(n_maps)
    maps = SEQUENTIAL_CMAPS[:n_maps]
    
    label_to_color = {}
    for label, agg in label_to_hierarchy.items():
        if len(agg) == 1:
            label_to_color[label] = sns.color_palette(maps[label], n_colors=1)[0]
        else:
            cm = maps[agg[0]]
            
            spot = agg[:-1]
            mask = (aggs[:, :-1] == spot).sum(-1) == len(spot)

            n_colors = np.count_nonzero(mask)

            labs = list(aggs[mask][:, -1])
            idx = labs.index(agg[-1])

            label_to_color[label] = sns.color_palette(cm, n_colors=n_colors)[idx]
    return label_to_color
            
        
```
